```python
import matplotlib.pyplot as plt
import numpy as np

# Set up the figure and axis
plt.figure(figsize=(8, 6))
ax = plt.gca()

# Small rectangle with red dots
small_rect_x = np.random.uniform(1, 2, 2000)  # X coordinates
small_rect_y = np.random.uniform(6, 7, 2000)  # Y coordinates
ax.scatter(small_rect_x, small_rect_y, color='gold', label='Patients suitable for clinical trial', s=2)

# Larger rectangle with mostly yellow dots and some red dots
large_rect_x = np.random.uniform(1, 5, 10000)  # X coordinates
large_rect_y = np.random.uniform(1, 5, 10000)  # Y coordinates
colors = ['lightgray'] * 8000 + ['gold'] * 2000  # 80% yellow, 20% red
np.random.shuffle(colors)  # Shuffle to mix red and yellow dots
ax.scatter(large_rect_x, large_rect_y, color=colors, label='Patients not suitable for clinical trial', s=2)

# Add labels and legend
ax.legend()

# Remove numbers on axes
ax.set_xticklabels([])
ax.set_yticklabels([])

# Show the plot
plt.show()
```
