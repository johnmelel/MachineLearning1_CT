```python
import pandas as pd
import pyarrow.parquet as pq
#import dask.dataframe as dd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
diagnoses_icd_df = pq.read_table('../Data/diagnoses_icd/').slice(0, 100)
diagnoses_icd_df = diagnoses_icd_df.to_pandas()
discharge_df = pq.read_table('../Data/discharge/').slice(0, 100)
discharge_df = discharge_df.to_pandas()

discharge_detail_df = pd.read_parquet('../Data/discharge_detail/')
drgcodes_df = pd.read_parquet('../Data/drgcodes/')
d_hcpcs_df = pd.read_parquet('../Data/d_hcpcs/')
d_icd_diagnoses_df = pd.read_parquet('../Data/d_icd_diagnoses/')
d_icd_procedures_df = pd.read_parquet('../Data/d_icd_procedures/')
d_labitems_df = pd.read_parquet('../Data/d_labitems/')
emar_df = pq.read_table('../Data/emar/').slice(0, 100)
emar_df = emar_df.to_pandas()

emar_detail_df = pq.read_table('../Data/emar_detail/').slice(0, 100)
emar_detail_df = emar_detail_df.to_pandas()

hcpcsevents_df = pq.read_table('../Data/hcpcsevents/').slice(0, 100)
hcpcsevents_df = hcpcsevents_df.to_pandas()

microbiologyevents_df = pq.read_table('../Data/microbiologyevents/').slice(0, 100)
microbiologyevents_df = microbiologyevents_df.to_pandas()

omr_df = pq.read_table('../Data/omr/').slice(0, 100)
omr_df = omr_df.to_pandas()

patients_df = pd.read_parquet('../Data/patients/')
pharmacy_df = pq.read_table('../Data/pharmacy/').slice(0, 100)
pharmacy_df = pharmacy_df.to_pandas()

prescriptions_df = pq.read_table('../Data/prescriptions/').slice(0, 100)
prescriptions_df = prescriptions_df.to_pandas()

procedures_icd_df = pq.read_table('../Data/procedures_icd/').slice(0, 100)
procedures_icd_df = procedures_icd_df.to_pandas()

radiology_df = pq.read_table('../Data/radiology/').slice(0, 100)
radiology_df = radiology_df.to_pandas()

radiology_detail_df = pd.read_parquet('../Data/radiology_detail/')
```

```python
# Calculate the fill rate of the 'dod' column
dod_fill_rate = patients_df['dod'].notnull().mean() * 100
print(f"Fill rate of 'dod' column: {dod_fill_rate:.2f}%")
```

```python
patients_df['anchor_age'] = pd.to_numeric(patients_df['anchor_age'], errors='coerce')
patients_df['anchor_year'] = pd.to_numeric(patients_df['anchor_year'], errors='coerce')

# Display basic statistics
print(patients_df.describe(include='all'),"\n")

# Check for missing values
print(patients_df.isnull().sum())

# Plot the distribution of anchor_age
plt.figure(figsize=(10, 6))
sns.histplot(patients_df['anchor_age'], bins=20, kde=True)
plt.title('Distribution of Anchor Age')
plt.xlabel('Anchor Age')
plt.ylabel('Frequency')
plt.show()

# Plot the distribution of gender
plt.figure(figsize=(10, 6))
sns.countplot(x='gender', data=patients_df)
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

# Plot the distribution of anchor_year
plt.figure(figsize=(10, 6))
sns.histplot(patients_df['anchor_year'], bins=20, kde=True)
plt.title('Distribution of Anchor Year')
plt.xlabel('Anchor Year')
plt.ylabel('Frequency')
plt.show()
```
